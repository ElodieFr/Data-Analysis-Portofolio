#### Question 3 : 
Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal puis en kg de protéines.
Vous ferez cela à partir de ces informations :
- Population de chaque pays ;
- Disponibilité alimentaire donnée pour chaque produit et pour chaque pays en kcal/personne/jour ;
- Disponibilité alimentaire en protéines donnée pour chaque produit et pour chaque pays en g/personne/jour.

```python
# Suppression de la chine, ajout d'une colonne pour identifier les produits d'origine végétale et animale 
# Concat des Dataframes animaux et vegetaux
vegetaux = vegetaux[vegetaux['Code zone'] != 351]
animaux = animaux[animaux['Code zone'] != 351]
vegetaux['type'] = 'vegetale'
animaux['type'] = 'animal'
df_animaux_vegetaux = pd.concat([vegetaux, animaux])
```

```python
# Pivot en gardant les colonnes Zone, Code zone, Produit, type. 
# Pivot sur Élément
df_animaux_vegetaux = df_animaux_vegetaux.pivot_table(index=['Zone','Code zone','Produit', 'type'], values='Valeur', columns='Élément')
df_animaux_vegetaux = df_animaux_vegetaux.rename_axis(None, axis=1)
df_animaux_vegetaux.reset_index(inplace=True)
df_animaux_vegetaux = df_animaux_vegetaux.rename(columns={'Zone': 'pays', 'Code zone': 'code_pays'})
```

```python
# Conversion des colonnes exprimés en millier de tonnes en kg
col = ['Autres utilisations (non alimentaire)', 'Aliments pour animaux', 'Disponibilité intérieure', 'Exportations - Quantité', 'Importations - Quantité', 'Nourriture', 'Pertes', 'Production', 'Semences', 'Traitement', 'Variation de stock']

for var_col in col:
    df_animaux_vegetaux[var_col] *= 1000000
df_animaux_vegetaux.head()
```

```python
# Merge avec Dataframe population et calcul de la disponibilité alimentaire en kcal et kg de protéine
df_animaux_vegetaux = pd.merge(df_animaux_vegetaux, population_totale[['pays','population']], on='pays', how='inner')
df_animaux_vegetaux['dispo_alim_kcal'] = df_animaux_vegetaux["Disponibilité alimentaire (Kcal/personne/jour)"] * df_animaux_vegetaux["population"] * 365
df_animaux_vegetaux['dispo_alim_kg_prot'] = df_animaux_vegetaux["Disponibilité de protéines en quantité (g/personne/jour)"] *0.001 * df_animaux_vegetaux["population"] * 365
df_animaux_vegetaux.head()
```

#### Question 4 : 
A partir de ces dernières informations, et à partir du poids de la disponibilité alimentaire (pour chaque pays et chaque produit), calculez pour chaque produit le ratio "énergie/poids", que vous donnerez en kcal/kg. Vous pouvez vérifier la cohérence de votre calcul en comparant ce ratio aux données disponibles sur internet, par exemple en cherchant la valeur calorique d'un oeuf.

##### Indication : La disponibilité alimentaire en kcal/personne/jour est calculée par la FAO en multipliant la quantité Nouriture (Food) par le ratio énergie/poids (en kcal/kg), puis en le divisant par la population du pays puis par 365. Ici, on vous demande juste de retrouver le ratio énergie/poids que la FAO a utilisé dans son calcul.

```python
df_ratio = df_animaux_vegetaux[['pays', 'Produit','dispo_alim_kcal', 'dispo_alim_kg_prot','Nourriture']].copy()
df_ratio.replace(0, np.nan, inplace=True)
df_ratio['ratio_kcal_kg'] = df_ratio['dispo_alim_kcal'] / df_ratio['Nourriture'] 
df_ratio['ratio_proteines_poids'] = df_ratio['dispo_alim_kg_prot'] / df_ratio['Nourriture']
df_ratio = df_ratio[['Produit','ratio_kcal_kg','ratio_proteines_poids']].groupby('Produit').mean()
df_ratio.reset_index(inplace=True)
df_ratio.head()
```

```python
# Vérification du calcul à partir de la valeur calorique d'un oeuf
# La valeur calorique d'un oeuf est d'environ 1 430 (kcal/kg)
valeur_calorique_oeuf = df_ratio[(df_ratio['Produit'] == 'Oeufs')]['ratio_kcal_kg'].mean()
print('{:.2f}'.format(valeur_calorique_oeuf))
```

```python
# Teneur en protéines de l'avoine
prot_avoine = df_ratio[(df_ratio['Produit'] == 'Avoine')]['ratio_proteines_poids'].mean()
print('{:.2f}'.format(prot_avoine))
```
